```python
print(__doc__)
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)

iris = datasets.load_iris()
a = iris.data
b = iris.target

estimators = [('k_means_iris_7', KMeans(n_clusters=7)),
              ('k_means_iris_4', KMeans(n_clusters=4)),
              ('k_means_iris_bad_init', KMeans(n_clusters=4, n_init=1,
                                               init='random'))]

figure_num = 1
titles = ['8 cl', '3 cl', '3 cl, Bad init']
for name, estimator in estimators:
    fig = plt.figure(figure_num, figsize=(4, 3))
    ax = Axes3D(fig, rect=[0, 0, .77, 1], elev=34, azim=194)
    estimator.fit(a)
    labels = estimator.labels_

    ax.scatter(a[:, 3], a[:, 0], a[:, 2],
               c=labels.astype(np.float), edgecolor='k')

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('Petal width')
    ax.set_ylabel('Sepal length')
    ax.set_zlabel('Petal length')
    ax.set_title(titles[figure_num - 1])
    ax.dist = 12
    figure_num = figure_num + 1


figure123 = plt.figure(figure_num, figsize=(4, 3))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=34, azim=194)

for name, label in [('Setosa', 0),
                    ('Versicolour', 1),
                    ('Virginica', 2)]:
    ax.text3D(a[b == label, 3].mean(),
              a[b == label, 0].mean(),
              a[b == label, 2].mean() + 2, name,
              horizontalalignment='center',
              bbox=dict(alpha=.2, edgecolor='w', facecolor='w'))


# 3D representation.
b = np.choose(b, [1, 2, 0]).astype(np.float)

ax.scatter(a[:, 3], a[:, 0], a[:, 2], c=b, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')
ax.set_title('Ground Truth')
ax.dist = 12

figure123.show()

```
