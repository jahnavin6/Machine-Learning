```python
#I have implemented the covariance operation. It will receive two arrays as parameters and will return the covariance(x,y) value. This might help in differentiating dependancies.
def covariance (u, v): 
    uhat=np.mean(u) 
    vhat=np.mean(v) 
    epsiQ=0
    for x,y in zip (u,v):
        epsiQ=epsiQ+(x-uhat)*(y-vhat) 
    return epsiQ/(len(u)-1)
```

```python
#Comparision of implemented function of the NumPy function
print (covariance ([3,5,7], [2,6,9]))

print (np.cov([3,5,7], [2,6,9]))

#I have taken the above values as test cases as one of the parameters is petal and sepal(2* criteria so raised the power by 2), other is the 3 types of iris (used the 3 multiples to test)
```

```python
def correlation (g, h):

    return(covariance(g,h)/(np.std(g,    ddof=1)*np.std(h,   ddof=1))) #ddof=1 is unbiased std
```
